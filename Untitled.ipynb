# Data split into Training and Testing Set

```python
df.drop(columns='date', axis=1, inplace=True)
```

```python
X=df.drop(columns='adj_close')
y = df['adj_close']
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y, test_size=0.2)
```

```python
print(f"X_train Data: {X_train.shape}")
print(f"y_train Data: {y_train.shape}")
print(f"X_test Data: {X_test.shape}")
print(f"y_test Data: {y_test.shape}")
```

# Feature Scaling

```python
scaler = StandardScaler()
scaler.fit(X_train,X_test)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Model 

# Linear R | 99.7 Accuracy

```python
le = LinearRegression()
le.fit(X_train_scaled,y_train)
linear_pred = le.predict(X_test_scaled)
```

```python
print(r2_score(y_test,linear_pred))
print(mean_squared_error(y_test, linear_pred))
```

# Random F | 99.7 Accuracy

```python
lr = RandomForestRegressor()
lr.fit(X_train_scaled,y_train)
random_pred = le.predict(X_test_scaled)
```

```python
print(r2_score(y_test,random_pred))
print(mean_squared_error(y_test, random_pred))
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test.values, label='Actual', color='black')
plt.plot(linear_pred, label='Linear R Prediction', color='blue', linestyle= '--')
plt.plot(random_pred, label='Random F  Prediction', color='tomato', linestyle= '--')
plt.title('Model Predictions',fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```
